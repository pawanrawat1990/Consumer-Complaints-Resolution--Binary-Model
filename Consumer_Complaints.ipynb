```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv("D:/Python/Edvancer/Project 1/Consumer_Complaints_train.csv")
```

```python
test=pd.read_csv("D:/Python/Edvancer/Project 1/Consumer_Complaints_test_share.csv")
```

```python
train_updated =train[[ col for col in train.columns if col != 'Consumer disputed?' ] + ['Consumer disputed?']]
```

```python
test["Consumer disputed?"]="0"
```

```python
train_updated["Temp"]= "train"
test["Temp"]="test"
```

```python
combi=pd.concat([train_updated,test],axis=0)
```

```python
combi.shape
```

```python
combi.dtypes
```

```python
import datetime
```

```python
combi["Date received"]=pd.to_datetime(combi["Date received"],format="%Y-%m-%d")
```

```python
combi["Date sent to company"]=pd.to_datetime(combi["Date sent to company"],format="%Y-%m-%d")
```

```python
combi["days_diff"]=combi["Date sent to company"]-combi["Date received"]
```

```python
combi["days_taken"]=combi["days_diff"].dt.days
```

```python
combi.drop(["Date received","Date sent to company","days_diff"],axis=1,inplace=True)
```

```python
for i in combi.select_dtypes(["object"]).columns:
    print(i,":",combi[i].nunique())
```

```python
combi.isnull().sum(axis=0)
```

```python
for col in ['Sub-product','Sub-issue','Consumer complaint narrative',
            'Company public response','Tags','Consumer consent provided?','ZIP code']:
    varname=col.replace('-','_').replace('?','').replace(" ",'_')+'_ismissing'
    combi[varname]=np.where(pd.isnull(combi[col]),1,0)
    combi.drop([col],1,inplace=True)
```

```python
combi["Submitted via"].value_counts(dropna=False)
```

```python
combi["Submitted via"].fillna("Web",inplace=True)
```

```python
combi['Timely response?']=np.where(combi['Timely response?']=="Yes",1,0)
```

```python
combi["State"].value_counts()
```

```python
k=combi["State"].value_counts()
for i in k.axes[0][0:15]:
    varname='State_'+i.replace(',','_').replace(' ','_')
    combi[varname]=np.where(combi['State']==i,1,0)
del combi["State"]
```

```python
combi['Company'].value_counts()
```

```python
k1=combi['Company'].value_counts()
for val in k1.axes[0][0:12]:
    varname='Company_'+val.replace(',','_').replace(' ','_')
    combi[varname]=np.where(combi['Company']==val,1,0)
del combi["Company"]
```

```python
dummy=pd.get_dummies(combi[["Product","Company response to consumer","Submitted via"]],drop_first=True)
```

```python
combi=pd.concat([combi,dummy],axis=1)
```

```python
combi_updated=combi.drop(["Product","Company response to consumer","Submitted via"],axis=1)
```

```python
##### Lower Case ######################################
combi_updated["Issue"]=combi_updated["Issue"].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
### remove special characters, numbers, punctuations #####
combi_updated["Issue"] = combi_updated["Issue"].str.replace("[^a-zA-Z#]", " ")
```

```python
#### Removing Short words ######
combi_updated["Issue"] = combi_updated["Issue"].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
combi_updated.drop(["Complaint ID"],axis=1,inplace=True)
```

```python
training=combi_updated.loc[combi_updated["Temp"]=="train"]
testing=combi_updated.loc[combi_updated["Temp"]=="test"]
```

```python
training.drop(["Temp"],axis=1,inplace=True)
testing.drop(["Temp","Consumer disputed?"],axis=1,inplace=True)
```

```python
training["Consumer disputed?"]=np.where(training["Consumer disputed?"]=="Yes",1,0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train=training.drop(["Consumer disputed?"],axis=1)
y_train=training["Consumer disputed?"]
```

```python
np.random.seed(324)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.30,stratify=y_train)
```

```python
from xgboost.sklearn import XGBClassifier
```

```python
clf=XGBClassifier(objective='binary:logistic',scale_pos_weight=3,n_estimators=200)
```

```python
clf.fit(X_train,Y_train)
```

```python
p=clf.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(Y_test,p)
```
